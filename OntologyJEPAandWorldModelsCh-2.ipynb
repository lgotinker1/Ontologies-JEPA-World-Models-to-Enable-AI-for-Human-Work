**2.1 What an Ontology Actually Is**

Ontologies are often confused with taxonomies, schemas, or databases. But an ontology is far more powerful.

An ontology is a **formal, explicit specification of the concepts, relationships, constraints, and rules that define a domain**.

Ontologies include:

- **Entities** (Patient, Package, RobotArm, Invoice)

- **Attributes** (weight, temperature, status, priority)

- **Relationships** (belongsTo, locatedAt, dependsOn, supervises)

- **Constraints** (a forklift cannot enter a restricted zone)

- **Rules** (a medication must match a prescription)

- **Actions** (PickUp, AdministerMedication, DeploySoftware)

- **Processes** (triage workflow, assembly sequence, routing pipeline)

Ontologies define *what exists* and *how it behaves*.

**2.1.1 Ontologies vs. Taxonomies**

A taxonomy is a hierarchy of categories.

An ontology is a **network of meaning**.

Taxonomy:

- “A dog is a mammal.”

Ontology:

- “A dog is a mammal that has an owner, requires vaccinations, can be walked, and cannot be left alone for more than 8 hours.”

Ontologies capture **relationships**, **constraints**, and **affordances**.

**2.1.2 Ontologies vs. Databases**

Databases store data.  
Ontologies store **meaning**.

Database:

- “Patient #123 has medication X.”

Ontology:

- “Medication X interacts with Medication Y, requires dosage Z, and must be administered by a licensed nurse.”

Ontologies encode **rules**, **semantics**, and **context**.

**2.1.3 Ontologies vs. Knowledge Graphs**

Knowledge graphs are *instances* of ontologies.

Ontology: the schema  
Knowledge graph: the data that fits the schema

Ontology:

- “A package has a weight.”

Knowledge graph:

- “Package #A123 weighs 4.2 kg.”

Ontologies define the **structure**; knowledge graphs fill it with **facts**.

**2.1.4 Ontologies vs. Machine Learning Models**

Machine learning models learn patterns.  
Ontologies define **meaning**.

ML model:

- “This image probably contains a cup.”

Ontology:

- “A cup is a container with volume, temperature, material, and affordances such as graspable, fillable, and movable.”

Ontologies give ML models a **semantic anchor**.
